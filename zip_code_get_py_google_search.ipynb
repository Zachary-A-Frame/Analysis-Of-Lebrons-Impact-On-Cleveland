### Cities/Towns in Cuyahoga County
* Will need to clean the data.

```python
import json
from api_key import g_key
import requests
import pandas as pd
from pprint import pprint

output_data_file = "zip_cities.csv"

zip_search = []
city_zip = []
city_name = []
city_type1 = []
city_type2 = []
city_county = []
city_state = []
city_lat = []
city_lng = []
zip_status = []
city_cnt = 0



# Testing Area to limit the calls.
zip_data = [44149, 44139]
#zip_data = [43069, 43070, 43071, 43072, 43073, 43076, 43077, 45365, 45367, 44139, 44134, 44116]



params = {
    'sensor' : "false",
    'key' : g_key
    
}

google_url = "https://maps.googleapis.com/maps/api/geocode/json?"

for item in zip_data:
    
    params["address"] = item
    data_response = requests.get(google_url, params=params)
    zip_json = data_response.json() 
    
    try:
        
        city_lat.append(zip_json["results"][0]["geometry"]["location"]["lat"])
        city_lng.append(zip_json["results"][0]["geometry"]["location"]["lng"])
        zip_status.append(zip_json["status"])
        #city_zip.append(zip_json["results"][0]["address_components"][0]["long_name"])
        #city_type1.append(zip_json["results"][0]["address_components"][0]["types"][0])
        
        found_zip = False
        found_city = False
        found_county = False
        found_state = False
        
        for row in zip_json['results']:
            
            for address in row['address_components']:
                
                #print("check 2")
                #print(address)
                
                if 'postal_code'  in address['types']:
                    if not found_zip:
                        city_zip.append(address["long_name"])
                        city_type1.append("postal_code")
                        found_zip = True
                
                if 'locality'  in address['types']:
                    if not found_city:
                        city_name.append(address["long_name"])
                        found_city = True
                
                if 'administrative_area_level_2' in address['types']:
                    if not found_county:
                        city_county.append(address["long_name"])
                        found_county = True

                if 'administrative_area_level_1'  in address['types']:
                    if not found_state:
                        city_state.append(address["short_name"])
                        found_state = True
                    
        zip_search.append(item)
        
        if not found_zip:
            city_zip.append(address["None"])
            city_type1.append("None")
        
        if not found_city: city_name.append("None")
        if not found_county: city_county.append("None")
        if not found_state: city_county.append("None")
                
                
                
 
    except:

        print(f"## Possible Zip Issue--> {item} <-- ##")
        pass

    
    print(f"## City Zip--> {item} and City Cnt--> {city_cnt}  ##")
    city_cnt = city_cnt + 1


cities_in_county_df = pd.DataFrame({
    'Zip Code': city_zip,
    'City Name': city_name,
    'City Type Zip': city_type1,
    'County': city_county,
    'State': city_state,
    'Lat': city_lat,
    'Lng': city_lng,
    'Zip Code Status': zip_status,
    'Zip Check' : zip_search
    })   
    
#print(json.dumps(zip_json, indent=4, sort_keys=False))


#put data in csv in case it is needed.
cities_in_county_df.to_csv(output_data_file, index= False)

cities_in_county_df

```
