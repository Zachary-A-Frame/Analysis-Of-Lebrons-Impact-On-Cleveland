```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

#importing csv

data = "Resources/Employment_Wages.csv"

# Read Purchasing File and store into Pandas data frame
employment = pd.read_csv(data)
employment
```

```python
#line graph
x_axis = employment['Year']
y_axis = employment['Annual']

plt.plot(x_axis, y_axis, marker='o', color='red', label = "Employment")
plt.legend(loc="best")
plt.grid(True)
plt.title("Annual Employment Numbers for Cuyahoga County")
plt.xlabel("Year")
plt.ylabel("Employment")
plt.show()
```
