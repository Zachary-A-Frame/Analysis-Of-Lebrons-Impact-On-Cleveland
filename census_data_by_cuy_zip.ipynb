### Start Tearing into Census Data
* Will need to clean the data.

```python
zip_data = []
csv_file = "cuy_cities-v2.csv"
cuy_cities = pd.read_csv(csv_file)

zip_data = cuy_cities["Zip Code"]

zip_data
```

```python
cen_year = []
cen_zip = []
cen_zip_employees = []
cen_zip_total_payroll = []
cen_zip_bus_cnt = []
census_geo_id = []





# Testing Area to limit the calls.
#zip_data = [44113]
#zip_data = [44113, 44149, 44139, 44017, 44136]
#zip_data = [43071, 43072, 43073, 43076, 43077, 45365, 45367, 44139, 44134, 44116,44149, 44139, 44017, 44136]


city_cnt = 0
#params = {
#    'sensor' : "false",
#    'key' : g_key
    
#}


#url = f"https://api.census.gov/data/2007/zbp?get=GEO_TTL,EMP,NAICS_TTL,ESTAB,PAYANN&for=zipcode:44113&key={census_key}"

#response = requests.get(url)
#print(response.url)
#print(response)

#data = response.json()
#print(json.dumps(data, indent=4, sort_keys=True))



for year in range(2000, 2018):
    c_year = str(year)
    for item in zip_data:
        zip = item
        url = f"https://api.census.gov/data/{year}/zbp?get=GEO_ID,EMP,ESTAB,PAYANN&for=zipcode:{zip}&key={census_key}"
        data = requests.get(url)
        print(url)
        #census_json = data.json()
        #print(c_year)
        #print(data_response.url)
        #print(data_response)
        #print(json.dumps(census_json, indent=4, sort_keys=True))
        #temp = census_json
        
        
        try:
            census_json = data.json()
            census_geo_id.append(census_json[1][0])
            cen_zip_employees.append(census_json[1][1])       
            cen_zip_bus_cnt.append(census_json[1][2])      
            cen_zip_total_payroll.append(census_json[1][3])      
            cen_zip.append(census_json[1][4])
            cen_year.append(c_year)

        except:

            print(f"## Possible Zip Issue--> {zip} <-- ##")
            pass

    
        print(f"## City Zip--> {zip} and City Cnt--> {city_cnt} and Year--> {c_year} {data} ##")
        city_cnt = city_cnt + 1


census_data_df = pd.DataFrame({
     'Zip Code': cen_zip,
     'Census Geo ID': census_geo_id,
     'Zip Payroll': cen_zip_total_payroll,
     'Zip Business Count': cen_zip_bus_cnt,
     'Zip Employee Count': cen_zip_employees,
     'Census Year': cen_year
     })   
    
#print(json.dumps(zip_json, indent=4, sort_keys=False))


#put data in csv in case it is needed.
output_data_file = "census_zip_cities.csv"
census_data_df.to_csv(output_data_file, index= False)

census_data_df

```
